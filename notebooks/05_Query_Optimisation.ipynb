## Baseline Operational Query Performance Investigation

The following investigation evaluates baseline query execution performance before optimisation techniques are applied within the NorthStar operational intelligence environment.

The objective is to measure initial operational query efficiency across semi-structured operational records in order to establish a comparative foundation for subsequent indexing and optimisation analysis.

```python
start_time = time.time()

baseline_results = list(
    cases_collection.find(
        {"delivery_status": "Failed"}
    )
)

end_time = time.time()

baseline_execution_time = end_time - start_time

print("Baseline Query Execution Time:")
print(round(baseline_execution_time, 6), "seconds")

print("\nTotal Failed Operational Cases:")
print(len(baseline_results))
```

## MongoDB Index Optimisation Implementation

The following optimisation process implements indexing strategies designed to improve operational query efficiency across the NorthStar operational intelligence environment.

```python
cases_collection.create_index(
    [("delivery_status", 1)]
)

print("Operational delivery status index created successfully.")
```

## Optimised Operational Query Performance Investigation

The following investigation evaluates operational query performance after index optimisation has been implemented within the NorthStar operational intelligence environment.

The objective is to measure optimisation impact by comparing indexed operational query execution efficiency against the previously established baseline performance investigation.

```python
start_time = time.time()

optimised_results = list(
    cases_collection.find(
        {"delivery_status": "Failed"}
    )
)

end_time = time.time()

optimised_execution_time = end_time - start_time

print("Optimised Query Execution Time:")
print(round(optimised_execution_time, 6), "seconds")

print("\nTotal Failed Operational Cases:")
print(len(optimised_results))
```
