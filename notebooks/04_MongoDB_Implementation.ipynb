## Loading Semi-Structured Operational Records

The following implementation loads selected NorthStar operational datasets that will be transformed into MongoDB document structures supporting integrated operational visibility and flexible event-based analytics.

```python
# Load operational datasets

complaints_df = pd.read_csv("/content/complaints.csv")
incidents_df = pd.read_csv("/content/incidents.csv")
events_df = pd.read_csv("/content/app_events.csv")

print("Operational datasets loaded successfully.")
```

## Customer Complaint Document Transformation

The following implementation transforms customer complaint records into MongoDB-compatible document structures capable of supporting flexible complaint tracking, operational visibility, and scalable customer-service analytics.

```python
# Transform complaint records into MongoDB documents

complaint_documents = complaints_df.to_dict(orient='records')

# Preview first complaint document
complaint_documents[0]
```

## Customer Complaint Document Insertion

The following implementation inserts transformed customer complaint documents into the MongoDB operational intelligence environment to support scalable document-based storage and operational querying.

```python
# Insert complaint documents into MongoDB collection

complaints_collection.insert_many(complaint_documents)

print("Complaint documents inserted successfully.")
```

## MongoDB Operational Query Investigation

The following implementation performs document-based operational queries within the MongoDB operational intelligence environment to support customer complaint investigation and operational visibility analysis.

```python
# Query high-severity customer complaints

high_severity_complaints = complaints_collection.find(
    {"severity": "High"}
)

# Display first 5 high-severity complaints
for complaint in list(high_severity_complaints)[:5]:
    print(complaint)
```

## MongoDB Aggregation Pipeline Investigation

The following implementation uses MongoDB aggregation pipelines to analyse customer complaint patterns and identify operational issue concentrations across the NorthStar environment.

```python
# Analyse complaint concentration by complaint type

complaint_type_analysis = complaints_collection.aggregate([
    {
        "$group": {
            "_id": "$complaint_type",
            "total_complaints": {"$sum": 1}
        }
    },
    {
        "$sort": {
            "total_complaints": -1
        }
    }
])

# Display aggregation results
for result in complaint_type_analysis:
    print(result)
```
